```python
import cv2
import matplotlib.pyplot as plt
#import numpy as np
#%matplotlib inline
```

```python
from skimage import io
photo = io.imread('a3.jpg')
plt.imshow(photo)
```

```python
# Load the cascade
face_cascade = cv2.CascadeClassifier ('haarcascade_frontalface_default.xml')     # Read the input image
```

```python
# Read the input image
img = cv2.imread('a3.jpg')
```

```python
# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
```

```python
# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)   
```

```python
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)       # BGR= blue,green,red
```

```python
plt.imshow(convertToRGB(img))
#cv2.waitKey()
```
